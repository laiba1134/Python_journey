```python
while True:
  year = int(input("Enter the year you want to check wether it is leap year or not: "))
  if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
      print(f"{year} is a leap year")
  else:
      print(f"{year} is not a leap year")
  choice = input("Do you want to check another year ?")
  if choice.lower() != "yes":
      break
```
